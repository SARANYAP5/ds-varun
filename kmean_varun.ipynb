np.random.uniform(0, 1, 1000) generates an array of 1000 random numbers uniformly distributed between 0 and 1.

```python
x= np.random.uniform(0,1,1000)
y= np.random.uniform(0,1,1000)
x
```

```python

# Create an empty DataFrame
df_xy = pd.DataFrame(columns=["x", "y"])

# Assign values to columns
df_xy.x = x
df_xy.y = y

print(df_xy)
```

```python
#apply scatter plot
df_xy.plot("x","y",kind="scatter")
#apply kmean clustering

#m=KMeans(n_clusters=5)
#m=m.fit(df_xy)

m=KMeans(n_clusters=5).fit(df_xy)
```

```python
df_xy.plot(x="x",y="y",kind="scatter",c=m.labels_,cmap=plt.cm.coolwarm_r)
```

 cmap=plt.cm.coolwarm_r

Argument: cmap defines the colormap to be used for coloring the points.

plt.cm.coolwarm_r is a reversed version (_r) of the coolwarm colormap from Matplotlib's colormap library. This colors the points according to their cluster labels.
